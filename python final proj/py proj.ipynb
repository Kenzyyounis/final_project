```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from pandas.api.types import is_numeric_dtype
```

```python
 pd.read_csv('data/Womens_Reviews.csv')
```

```python
class DataProcessor:
    def load_data(self):
        data = pd.read_csv('data/Womens_Reviews.csv')
        return data
    
    def clean_data(self, data):
        # Remove rows with missing values
        data.isnull()
        return data.dropna()
data__pro = DataProcessor()
df = data__pro.load_data()
cleaned_data = data__pro.clean_data(df)
```

```python
cleaned_data
```

```python
class DataAnalyzer:
    def __init__(self, cleaned_data):
        self.cleaned_data = cleaned_data
        
    def get_average_of_column(self, column):
        if is_numeric_dtype(cleaned_data[column]) == True:
            return self.cleaned_data[column].mean()
        else:
            print("Enter a numeric column")
    def get_distribution_of_column(self, column):
        if is_numeric_dtype(cleaned_data[column]) == True:
            return self.cleaned_data[column].std()
        else:
            print("Enter a numeric column")

    def get_median_of_column(self, column):
        if is_numeric_dtype(cleaned_data[column]) == True:
            return self.cleaned_data[column].median()
        else:
            print("Enter a numeric column")

    def get_mode_of_column(self, column):
        if is_numeric_dtype(cleaned_data[column]) == True:
            return self.cleaned_data[column].mode()
        else:
            print("Enter a numeric column")
```

```python
class DataVisualizer:
    def __init__(self, cleaned_data):
        self.cleaned_data = cleaned_data
        
    def plot_line_chart(self,X, column, title):
        sns.lineplot(x=X, y=column, data=self.cleaned_data)
        h= int(input("Enter 1 if you want to save this figure or Enter 0 if you don't want to save it: "))
        while h>1 or h<0:
            h= int(input("Please enter 0 or 1: "))
        if(h==1):
            plt.title(title)  
            plt.savefig("Output/"+title+".png")
        plt.show()
        
    def plot_distribution(self,Y, column, title):
        sns.barplot(x=column,y=Y, data=self.cleaned_data)
        h= int(input("Enter 1 if you want to save this figure or Enter 0 if you don't want to save it: "))
        while h>1 or h<0:
            h= int(input("Please enter 0 or 1: "))
        if(h==1):
            plt.title(title)  
            plt.savefig("Output/"+title+".png")
        plt.show()
        
    def plot_pie(self, column, title):
        myLabels = self.cleaned_data[column].unique()
        values = [self.cleaned_data[column].value_counts()[x] for x in myLabels]
        plt.pie(x=values, data=self.cleaned_data, labels = myLabels)
        h= int(input("Enter 1 if you want to save this figure or Enter 0 if you don't want to save it: "))
        while h>1 or h<0:
            h= int(input("Please enter 0 or 1: "))
        if(h==1):
            plt.title(title)  
            plt.savefig("Output/"+title+".png")
        plt.show()
        
    def plot_scatter(self, column_x, column_y, title):
        sns.scatterplot(x=column_x, y=column_y, data=self.cleaned_data)
        h= int(input("Enter 1 if you want to save this figure or Enter 0 if you don't want to save it: "))
        while h>1 or h<0:
            h= int(input("Please enter 0 or 1: "))
        if(h==1):
            plt.title(title)  
            plt.savefig("Output/"+title+".png")
        plt.show()
```

```python
data_processor = DataProcessor()
data = data_processor.load_data()
cleaned_data = data_processor.clean_data(data)
data_analyzer = DataAnalyzer(cleaned_data)
data_visualizer = DataVisualizer(cleaned_data)
```

```python
average_ratings = data_analyzer.get_average_of_column('Age')
rating_distribution = data_analyzer.get_distribution_of_column('Age')
recommendation_distribution = data_analyzer.get_median_of_column('Age')
feedback_count_distribution = data_analyzer.get_mode_of_column('Age')
```
